```python
import pandas as pd
from scipy.stats import pearsonr
import re
import os

df = pd.read_csv('titanic/train.csv', index_col='PassengerId')

def save_answer(filename, answer_str):
    with open(filename, 'w', encoding='utf-8') as f:
        f.write(str(answer_str).strip())
    print(f"Ответ для {filename} успешно сохранен: '{answer_str}'")

sex_counts = df['Sex'].value_counts()
save_answer('1.txt', f"{sex_counts['male']} {sex_counts['female']}")

save_answer('2.txt', f"{df['Survived'].mean() * 100:.2f}")

save_answer('3.txt', f"{(df['Pclass'] == 1).mean() * 100:.2f}")

save_answer('4.txt', f"{df['Age'].mean():.2f} {df['Age'].median():.2f}")

corr, _ = pearsonr(df['SibSp'], df['Parch'])
save_answer('5.txt', f"{corr:.2f}")

def get_female_first_name(full_name):
    full_name = full_name.strip()
    
    if "Mrs." in full_name and "(" in full_name:
        brackets_content = re.search(r'\((.*?)\)', full_name)
        if brackets_content:
            inner_name = brackets_content.group(1).replace('"', '').strip()
            return inner_name.split()[0]
            
    match = re.search(r'(?:Miss\.|Mrs\.|Mlle\.|Mme\.)\s+([A-Za-z]+)', full_name)
    if match:
        return match.group(1)
    return None

female_names = df[df['Sex'] == 'female']['Name']
extracted_names = female_names.apply(get_female_first_name).dropna()
save_answer('6.txt', extracted_names.value_counts().idxmax())

for file in [f"{i}.txt" for i in range(1, 7)]:
    if os.path.exists(file):
        os.remove(file)
        print(f"Файл {file} удалён.")
    else:
        print(f"Файл {file} не найден.")
```
