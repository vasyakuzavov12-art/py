```python
import pandas as pd
import numpy as np
from sklearn.metrics import roc_auc_score

df = pd.read_csv('logistic.csv', header=None)
y = df[0].values
X = df[[1, 2]].values

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def gradient_descent(X, y, C=0.0, k=0.1, max_iter=10000, eps=1e-5):
    w1, w2 = 0.0, 0.0
    l = len(y)
    
    for i in range(max_iter):
        margin = y * (X[:, 0] * w1 + X[:, 1] * w2)
        grad_coef = 1.0 - 1.0 / (1.0 + np.exp(-margin))
        
        sum_w1 = np.sum(y * X[:, 0] * grad_coef)
        sum_w2 = np.sum(y * X[:, 1] * grad_coef)
        
        w1_new = w1 + k * (1.0 / l) * sum_w1 - k * C * w1
        w2_new = w2 + k * (1.0 / l) * sum_w2 - k * C * w2
        
        if np.sqrt((w1_new - w1)**2 + (w2_new - w2)**2) <= eps:
            print(f"Конвергенция достигнута на шаге {i} (C={C})")
            return w1_new, w2_new
            
        w1, w2 = w1_new, w2_new
        
    print(f"Достигнуто максимальное число итераций {max_iter} (C={C})")
    return w1, w2

w1_no_reg, w2_no_reg = gradient_descent(X, y, C=0.0)
pred_no_reg = sigmoid(X[:, 0] * w1_no_reg + X[:, 1] * w2_no_reg)
auc_no_reg = roc_auc_score(y, pred_no_reg)

w1_reg, w2_reg = gradient_descent(X, y, C=10.0)
pred_reg = sigmoid(X[:, 0] * w1_reg + X[:, 1] * w2_reg)
auc_reg = roc_auc_score(y, pred_reg)

answer = f"{auc_no_reg:.3f} {auc_reg:.3f}"

with open('logistic_answer.txt', 'w', encoding='utf-8') as f:
    f.write(answer.strip())

print("\n--- Итоговые результаты ---")
print(f"AUC-ROC без регуляризации: {auc_no_reg:.3f}")
print(f"AUC-ROC с регуляризацией (C=10): {auc_reg:.3f}")
print(f"Строка ответа: '{answer}'")
```
