```python
text = "Введите русскоязычный текст: "
new_text = text.replace('а', 'о').replace('А', 'О')
replace_count = text.count('а') + text.count('А')
total_chars = len(text)
print(f"Измененная строка: {new_text}")
print(f"Количество замен: {replace_count}")
print(f"Общее количество символов: {total_chars}")
```

```python
import math

def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

A, B, C, D = 2, 3, 4, 5
result_num = A * D
result_den = B * C
gcd_val = gcd(result_num, result_den)
result_num //= gcd_val
result_den //= gcd_val
print(f"{A}/{B} / {C}/{D} = {result_num}/{result_den}")

print("-" * 30)

def is_inside_circle(x, y, a, b, R):
    return (x - a) ** 2 + (y - b) ** 2 < R ** 2

a, b, R = 0, 0, 5
points = [(3, 4), (-3, 4), (6, 0), (0, 0), (1, 1), (5, 0)]
inside_count = sum(1 for x, y in points if is_inside_circle(x, y, a, b, R))

print(f"Окружность (x-{a})^2 + (y-{b})^2 = {R}^2")
print(f"Точки: {points}")
print(f"Количество точек внутри окружности: {inside_count}")
```

```python
matrix = [
    [5, 2, 3],
    [4, 8, 6],
    [1, 0, 7]
]

row_sums = [sum(row) for row in matrix]
max_sum = max(row_sums)
min_sum = min(row_sums)
max_row_index = row_sums.index(max_sum)
min_row_index = row_sums.index(min_sum)

print("Матрица:")
for row in matrix:
    print(row)

print(f"\nСтрока с наибольшей суммой элементов:")
print(f"Строка {max_row_index + 1}: {matrix[max_row_index]}, Сумма: {max_sum}")

print(f"\nСтрока с наименьшей суммой элементов:")
print(f"Строка {min_row_index + 1}: {matrix[min_row_index]}, Сумма: {min_sum}")

print("-" * 30)

A = [
    [5, -2, 3],
    [-4, 8, -6],
    [1, -7, 4]
]

N = len(A)
for i in range(N):
    for j in range(N):
        if A[i][j] < 0:
            A[i][j] = 0
        elif A[i][j] > 0:
            A[i][j] = 1

print("Нижняя треугольная матрица:")
for i in range(N):
    for j in range(N):
        if i >= j:
            print(A[i][j], end=" ")
        else:
            print(" ", end=" ")
    print()
```
