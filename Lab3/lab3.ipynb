```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier

def load_data(filepath, index_col='PassengerId'):
    return pd.read_csv(filepath, index_col=index_col)

def prepare_features(df, features, target):
    X = df[features]
    y = df[target]
    X_clean = X.dropna()
    y_clean = y.loc[X_clean.index]
    X_clean = X_clean.copy()
    X_clean['Sex'] = X_clean['Sex'].map({'male': 1, 'female': 0})
    return X_clean, y_clean

def train_tree(X, y, random_state=241):
    clf = DecisionTreeClassifier(random_state=random_state)
    clf.fit(X, y)
    return clf

def get_top_features(importances, feature_names, n_top=2):
    return pd.Series(importances, index=feature_names).sort_values(ascending=False).index[:n_top]

df = load_data('titanic/train.csv')
features = ['Pclass', 'Fare', 'Age', 'Sex']

X_clean, y_clean = prepare_features(df, features, 'Survived')
clf = train_tree(X_clean, y_clean)

feature_importances = pd.Series(clf.feature_importances_, index=features)
print("Важность всех признаков:")
print(feature_importances)

top_two = get_top_features(clf.feature_importances_, features)
answer = ",".join(top_two)

with open('titanic_tree_answer.txt', 'w', encoding='utf-8') as f:
    f.write(answer.strip())

print(f"\nОтвет успешно записан в файл: '{answer}'")
```
